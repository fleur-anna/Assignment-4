```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define a function to calculate the probability
def P(N):
    
    #need to define what to return for N = 1 and N = 2, since the formula requires two previous values
    if N == 1:
        return 1
    elif N == 2:
        return 0.5
    
    #else use the formula in conjecture 1
    else:
        return 0.5*P(N-1) + 0.5*P(N-2)

#find P(25)
print('P(25) = ', P(25))
```

```python
#printing P(1),...,P(5) to check whether they are the same as those calculated by hand
print('P(1) = ', P(1))
print('P(2) = ', P(2))
print('P(3) = ', P(3))
print('P(4) = ', P(4))
print('P(5) = ', P(5))
```

```python
#plotting the probabulities for increasing N
Prob = np.zeros(40)
x = np.zeros(40)
compare1 = np.zeros(40)

for i in range(1, 40):
    Prob[i] = P(i)
    x[i] = i
    
for j in range(40):
    compare1[j] = 2/3
    
plt.plot(x, Prob, label = 'P(N)', linewidth = 2)
plt.plot(x, compare1, '--', label = r'$\frac{2}{3}$', linewidth = 2)
plt.xlabel('N')
plt.ylabel('Probability')
plt.title('P(N) as N increases, for numbers 1 and 2 on Coin')
plt.legend(fontsize = 'large')
```

```python
#doing the same as before but for the new formula for 1 and 3 on the coin
def P2(N):
    if N == 1:
        return 1
    elif N == 2:
        return 0.5
    elif N == 3: 
        return 0.25
    else:
        return 0.5*P2(N-1) + 0.5*P2(N-3)


Prob2 = np.zeros(40)
x2 = np.zeros(40)
compare2 = np.zeros(40)

for i in range(1, 40):
    Prob2[i] = P2(i)
    x2[i] = i
    
for j in range(40):
    compare2[j] = 0.5
    
plt.plot(x2, Prob2, label = 'P(N)', linewidth = 2)
plt.plot(x2, compare2, '--', label = '0.5', linewidth = 2)
plt.xlabel('N')
plt.ylabel('Probability')
plt.title('P(N) as N increases, for numbers 1 and 3 on coin')
plt.legend()
```
